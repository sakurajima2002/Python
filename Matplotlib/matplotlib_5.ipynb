```python
#Presentacion del analisis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_crudo = pd.read_csv(r'/home/rei/Documentos/Python/dow_jones_index.data')
df_crudo.head()
```

```python
""" El parametro converters de la funcion pd.read_csv nos permite
aplicar una operacion en las columnas """

columnas = ['accion', 'fecha', 'precio']
df = pd.read_csv(r'/home/rei/Documentos/Python/dow_jones_index.data',
                usecols=[1, 2, 6],
                header=0,
                names = columnas,
                parse_dates= ['fecha'],
                converters= {'precio':
                            lambda x:
                            float(x.replace('$', ''))
                            }
                )

df
```

```python
#¿Que periodo en el tiempo representan los datos?
df.fecha.dt.year.value_counts()
```

```python
#Meses a los que corresponden los datos
df.fecha.dt.month_name().value_counts()
```

```python
#En el mes de abril hay 30 registros ,as que en el resto de meses ¿por que?
df['mes'] = df.fecha.dt.month_name()
filtro1 = df['mes'] == 'April'
filtro2 = df['mes'] == 'May'

df[filtro1].groupby('accion')['mes'].value_counts()
```

```python
df[filtro2].groupby('accion')['mes'].value_counts()
```

```python
#¿Cuales acciones en promedio tuvieron la mayor y menor cotizacion?
df.groupby('accion').agg({'precio': [np.mean, np.std]}).sort_values([('precio', 'mean')], ascending=False)
```

```python
#¿Cuales acciones tuvieron la mayor y menor variacion de precio?
df.groupby('accion').agg({'precio': [np.mean, np.std]}).sort_values([('precio', 'std')], ascending=False)
```

```python
#pivot_table() nos genera un nuevo DataFrame reorganizado con la forma que deseamos
df_nuevo = df.pivot_table(index='fecha',
                        columns='accion',
                        values='precio'
                        )

df_nuevo
```

```python
#Ahora podemos centrarnos en estudiar el comportamiento en el tiempo de cada accion
fig, (pan1, pan2) = plt.subplots(2, 1, sharex=True, constrained_layout=True)

pan1.plot(df_nuevo.index, df_nuevo.CAT,
        color = "blue",
        marker = "o",
        linestyle = "solid",
        label = "CAT"
        )

pan2.plot(df_nuevo.index, df_nuevo.AA,
        color = "orange",
        marker = "v",
        linestyle = "solid",
        label = "AA"
        )

for pan in fig.get_axes():
        pan.set(xlabel="Fecha",
                ylabel="Dolares Americanos",
                title="Precio de la accion (al cierre)"
                )

pan1.legend()
pan2.legend()
```

```python
#¿Existe un correlacion entre los precios de CAT y el resto de las acciones?
df_nuevo.corr()
```

```python
df_nuevo.corr().loc['CAT']
```

```python
#Correlaciones positivas y altas de CAT con otras
corr_pos =df_nuevo.corr().loc['CAT'][df_nuevo.corr().loc['CAT'] > 0.8].sort_values(ascending=False)
corr_pos
```

```python
#Graficar Resultado
valores = corr_pos[1:5]

#index = eje x , values -> eje y
plt.bar(valores.index, valores.values, color='forestgreen')
plt.plot()

plt.xlabel("Acciones")
plt.ylabel("Coeficiente de correlacion")
plt.title("Correlaciones entre CAT y el resto")
plt.show()
```

```python
#Correlacion Negativa con CAT
corr_neg = df_nuevo.corr().loc['CAT'][df_nuevo.corr().loc['CAT'] < 0].sort_values(ascending=True)
corr_neg
```

```python
#Graficar Correlaciones negativas
valores = corr_neg[:]

plt.bar(valores.index, valores.values, color='brown')
plt.plot()

plt.xlabel('Acciones')
plt.ylabel('Coeficiente de correlacion')
plt.title('Correlaciones entre CAT y el resto')

plt.show()
```
