```python
#Tipos de Graficos
import numpy as np
import matplotlib.pyplot as plt

#Graficos de Lineas
aÃ±o = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
PIB = [300.2, 541.3, 1075.9, 2862.5, 5979.6, 10200.7, 14958.3]

plt.plot(aÃ±o, PIB, color='green', marker='o', linestyle='solid')
plt.title('PIB Nominal')

plt.ylabel('Millones de $')
plt.xlabel('AÃ±o')
```

```python
#Scatter plot
x1 = [3, 5, 7]
y1 = [3, 5, 9]

x2 = [1, 2, 3, 4]
y2 = [2, 3, 2, 3]

x3 = [8, 7, 2, 5, 4]
y3 = [6, 8, 7, 8, 7]

plt.scatter(x1, y1)
plt.scatter(x2, y2, marker="v", color="r")
plt.scatter(x3, y3, marker="^", color="m")

plt.title("Ejemplo Scatter plot")
plt.show()
```

```python
#Graficos de Barras
x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 7, 4, 7, 8, 3]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]

plt.bar(x1, y1, label="Barra azul", color='b')
plt.bar(x2, y2, label="Barra verde", color='g')

plt.plot()

plt.xlabel("Nro Barras")
plt.ylabel("Altura Barras")
plt.title("Ejemplo de Grafico de Barras")
plt.legend()
plt.show()
```

```python
#Histogramas
y = 5 + np.random.randn(100)
x = [e for e in range(len(y))]

plt.bar(x, y)

plt.title("Datos en crudo")
plt.show()
```

```python
plt.hist(y, bins=20) #bins -> Cantidad de columnas
plt.title("Historograma")
plt.show()
```

```python
plt.hist(y, cumulative=True, bins=20)

plt.title("Histograma acumulado")
plt.show()
```

```python
#Graficos Apilados
x = [1, 2, 3, 4, 5]
y1 = [1, 1, 2, 3, 5]
y2 = [0, 4, 2, 6, 8]
y3 = [1, 3, 5, 7, 9]

etiquetas = ['Sector Privado', 'Sector Publico', 'Sector Informal']

fig, panel = plt.subplots()

panel.stackplot(x, y1, y2, y3,
            labels=etiquetas,
            colors=['r','g','b'])

panel.legend(loc='upper left')

panel.set_title("Grafico Apilado")
plt.show()
```

```python
#Grafico Tortas
etiquetas = 'Financiero', 'Salud', 'Contruccion'
sectores = [56, 66, 24]
colores = ['c', 'y', 'r']

plt.pie(sectores, labels=etiquetas, colors=colores,
        startangle=0, # donde estara ubicado el primer espacio
        explode= (0, 0.2, 0), #separar un pedazo la torta para destacarlo
        autopct= '%1.2f%%', #valor dentro de la torta un digito con 2 decimales
        shadow=True,
        )

plt.axis('equal') # distribuya en distancias iguales en el eje
plt.title('Ejemplo de pie chart')
plt.show()
```

```python
#Boxplot
y1 = np.random.normal(100, 10, 100)
y2 = np.random.normal(80, 30, 100)
y3 = np.random.normal(90, 20, 100)
y4 = np.random.normal(70, 25, 100)

datos = [y1, y2, y3, y4]

fig, panel = plt.subplots()

panel.set_xticklabels(['Grupo_1', 'Grupo_2', 'Grupo_3', 'Grupo_4'])

bp = panel.boxplot(datos, patch_artist=True) #patch_artist -> boxplot coloreado

plt.title('Ejemplo Boxplot')

#Elementos Bloxblot
for box in bp['boxes']:
    box.set(color='r', linewidth=2)
    box.set(facecolor='b')

for whisker in bp['whiskers']:
    whisker.set(color='g', linewidth=2)

for cap in bp['caps']:
    cap.set(color='black', linewidth=2)

for median in bp['medians']:
    median.set(color='#b2df8a', linewidth=2)

for flier in bp['fliers']:
    flier.set(marker='*', color='r', alpha=1)
```
