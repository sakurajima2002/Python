```python
#Leyendas, etiquetas de eje y titulos
import numpy as np
import matplotlib.pyplot as plt

fig, pan_1  = plt.subplots(1, 1)

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)

pan_1.plot(x1, y1, label='Seno')
pan_1.legend(loc='best')
```

```python
#Leyendas MUltiples
fig, pan_1 = plt.subplots(1, 1)

pan_2 = pan_1.twinx()

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)

funcion_1 = pan_1.plot(x1, y1, color='b', marker='^',
            markeredgecolor='b', markerfacecolor='g',
            label='Seno')

x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)

funcion_2 = pan_2.plot(x2, y2, color='r', marker='o', 
            markeredgecolor='r', markerfacecolor='y',
            label='Coseno')

funciones = funcion_1 + funcion_2

labels = [f.get_label() for f in funciones]
plt.legend(funciones, labels, loc='best')

```

```python
#Etiquetas de Eje
fig, pan_1 = plt.subplots(1, 1)
pan_2 = pan_1.twinx()
x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)
funcion_1 = pan_1.plot(x1, y1, color='b', marker='^',
            markeredgecolor='b', markerfacecolor='g',
            label='Seno')

x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)
funcion_2 = pan_2.plot(x2, y2, color='r', marker='o', 
            markeredgecolor='r', markerfacecolor='y',
            label='Coseno')

funciones = funcion_1 + funcion_2
labels = [f.get_label() for f in funciones]
plt.legend(funciones, labels, loc='best')

pan_1.set_xlabel('$x$')
pan_1.set_ylabel('$y_1$')
pan_2.set_ylabel('$y_2$')
```

```python
#Titulo
fig, pan_1 = plt.subplots(1, 1)
pan_2 = pan_1.twinx()
x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)
funcion_1 = pan_1.plot(x1, y1, color='b', marker='^',
            markeredgecolor='b', markerfacecolor='g',
            label='Seno')

x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)
funcion_2 = pan_2.plot(x2, y2, color='r', marker='o', 
            markeredgecolor='r', markerfacecolor='y',
            label='Coseno')

funciones = funcion_1 + funcion_2
labels = [f.get_label() for f in funciones]
plt.legend(funciones, labels, loc='best')

pan_1.set_xlabel('$x$')
pan_1.set_ylabel('$y_1$')
pan_2.set_ylabel('$y_2$')

plt.title('Seno y Coseno')
plt.tight_layout() #Reducir el espacio figura
```
