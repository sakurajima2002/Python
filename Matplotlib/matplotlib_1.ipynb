```python
#Figuras y paneles
import numpy as np
import matplotlib.pyplot as plt

#subplots -> subgraficos
fig, panel = plt.subplots()
#ejes 
x = np.linspace(0, 4 * np.pi, 100) #linespace(valorinicio, valorfinalizacion, rango)
y = np.sin(x)

panel.plot(x, y)
```

```python
#Crear figura con paneles horizontales
fig, (pan_1, pan_2) = plt.subplots(2, 1) # subplots(filas, columnas)

#figura 1
x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)

pan_1.plot(x1, y1)

#figura 2
x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)

pan_2.plot(x2, y2)

#Guardar figura
plt.savefig(r'/home/rei/Documentos/Python/sin_cos.png')
```

```python
#crear la figura con dos paneles verticales
fig, (pan_1, pan_2) = plt.subplots(1, 2, constrained_layout = True)

#figura 1
x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)
pan_1.plot(x1, y1)

#figura 2
x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)

pan_2.plot(x2, y2)
```

```python
#Mas de dos graficos
fig, ((pan_1, pan_2), (pan_3, pan_4)) = plt.subplots(2, 2, constrained_layout=True)

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

pan_1.plot(x, y)
pan_2.plot(x, y**2)
pan_3.plot(x, -y)
pan_4.plot(x, -y**2)

for ac in fig.get_axes():
    ac.set(xlabel='x.label', ylabel='y.label')
```

```python
#sharex -> Permite indicar que los paneles compartan en el eje x

fig, (pan_1, pan_2) = plt.subplots(2, sharex=True)

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

pan_1.plot(x, y, 'tab:orange')
pan_2.plot(x + 1, -y, 'green')
```

```python
#sharey -> Permite indicar que los paneles compartan en el eje y

fig, (pan_1, pan_2) = plt.subplots(1, 2, sharey=True)

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

pan_1.plot(x, y)
pan_2.plot(-x, y+1)
```

```python
#Multiples paneles compartan ejes
fig, paneles = plt.subplots(2, 2, sharex='col', sharey='row')

(pan_1, pan_2), (pan_3, pan_4) = paneles

pan_1.plot(x, y)
pan_2.plot(x, y**2, 'tab:orange')
pan_3.plot(x + 1, -y, 'tab:green')
pan_4.plot(x + 2, -y**2, 'tab:red')
```

```python
#.twinx() -> graficar ambas funciones en un mismo panel
fig, pan_1 = plt.subplots(1, 1)
pan_2 = pan_1.twinx()

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)
pan_1.plot(x1, y1, 'y')

x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)
pan_1.plot(x2, y2, 'r')

```
